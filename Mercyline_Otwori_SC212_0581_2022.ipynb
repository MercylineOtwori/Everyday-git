```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)
X_poly_train = poly.fit_transform(Xh_train)
X_poly_test = poly.transform(Xh_test)

lr_poly = LinearRegression()
lr_poly.fit(X_poly_train, yh_train)
y_pred_poly = lr_poly.predict(X_poly_test)

print("Polynomial Regression R2:", r2_score(yh_test, y_pred_poly))
```

3)Ridge and Lasso


```python
from sklearn.linear_model import Ridge, Lasso

ridge = Ridge(alpha=1.0)
ridge.fit(Xh_train, yh_train)
ridge_pred = ridge.predict(Xh_test)

lasso = Lasso(alpha=0.1)
lasso.fit(Xh_train, yh_train)
lasso_pred = lasso.predict(Xh_test)

print("Ridge R2:", r2_score(yh_test, ridge_pred))
print("Lasso R2:", r2_score(yh_test, lasso_pred))
```
